```python
import pandas as pd
blood_df = pd.read_csv("transfusion.data")
blood_df.head()
```

```python
blood_df.isnull().sum()
```

```python
blood_df.describe()
```

```python
blood_df['CC per donation'] = blood_df['Monetary (c.c. blood)']/blood_df['Frequency (times)']
blood_df.head(20)
```

```python
blood_df['Donation per month'] = blood_df['Frequency (times)']/(blood_df['Time (months)'] - blood_df['Recency (months)'])
blood_df.head()
```

```python
blood_df.describe()
```

```python
blood_df =blood_df[blood_df["Recency (months)"]<blood_df['Time (months)']]
```

```python
blood_df.rename(columns = { 'whether he/she donated blood in March 2007':'donated in 2007'}, inplace = True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(blood_df['Frequency (times)'],blood_df['donated in 2007'])
```

```python
blood_df['donated in 2007'].value_counts(normalize=True).round(3)
```

```python
X = blood_df.drop("donated in 2007", axis = 1)
X.head()
```

```python
y = blood_df['donated in 2007']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 10)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators = 60)
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(solver = 'liblinear')
model_lr.fit(X_train, y_train)
model_lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
import numpy as np
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)
mean_score = np.mean(cross_val_score(LogisticRegression(solver = 'liblinear'),X,y,cv = cv))
mean_score
```
